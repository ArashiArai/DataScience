```python
from scipy import stats
import numpy as np
```

```python
group_a = [200,210,190,205,195]
group_b = [250,240,260,230,255]

t_stat, p_val = stats.ttest_ind(group_a,group_b)
print(f"P-Value: {p_val:.4f}")

if p_val < 0.05:
    print("The discount significantly increass sales")
else:
    print("The difference might just be luck")
```
