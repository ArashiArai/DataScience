```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame({
    "Square_Foot":[1500,2000,1200,2500,1800,1600,2200,1400,1900,2100,2800,1300,1700,2400,1100],
    'Bedrooms':[3,4,2,4,3,3,4,2,3,4,5,2,3,4,2],
    'House_Age':[10,5,20,2,15,8,1,12,7,4,1,25,11,3,30],
    'Price':[300,450,200,550,350,320,480,280,380,420,600,180,340,500,150]
})

df.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.title("correlations")
plt.show()
```

```python
X = df[["Square_Foot","Bedrooms","House_Age"]]
Y = df[["Price"]]

Model = LinearRegression().fit(X,Y)
```

```python
print(Model.coef_)
```

```python
print(f"{Model.intercept_[0]:.2f}")
```

```python
predicted = Model.predict([[1000,2,1]])
print(predicted[0][0])
```
